```python
import pandas as pd


crashes_df = pd.read_csv("Resources\Road_Crashes_for_five_Years_Victoria.csv")
crashes_df.head()
```

```python
crashes_cleaned_df = crashes_df.drop(columns=["X", "Y", "SRNS", "SRNS_ALL", "PED_CYCLIST_13_18", "PED_CYCLIST_5_12", "UNKNOWN", "PEDESTRIAN", "DRIVER", "PASSENGER", "BICYCLIST", "VICGRID_X", "VICGRID_Y", "NODE_ID", "OBJECTID", "ACCIDENT_NO", "ABS_CODE"])
```

```python
crashes_cleaned_df.head()
```

```python
crashes_cleaned_df.info()
```

```python
# Calculate Pearson correlation coefficients for quantitative data
correlation_matrix = crashes_cleaned_df.corr(method='pearson')

# Display the correlation matrix
print(correlation_matrix)
```

```python
import scipy.stats

# calculation of p-values for qualitative data

# Specify the two qualitative columns you want to analyze
column1 = crashes_cleaned_df['LIGHT_CONDITION']
column2 = crashes_cleaned_df['SEVERITY']

# Drop rows with NaN values in either of the columns
# crashes_cleaned_df.dropna(subset=['LIGHT_CONDITION', 'SEVERITY'], inplace=False)

# crashes_cleaned_df.info()


# Create a contingency table
contingency_table = pd.crosstab(column1, column2)

# Calculate the chi-squared statistic and p-value
chi2, p, _, _ = scipy.stats.chi2_contingency(contingency_table)

# print("Chi-squared statistic:", chi2)
print("p-value:", p)

```
